# Testing the fit metric calculation on the PISA dataset, comparing to the KST library by Milan Segedinac

1. Imporitng both libraries and requirements

```python
!git clone https://github.com/Alexe1900/iita_python.git
!git clone https://github.com/milansegedinac/kst.git
!pip install numpy
!pip install pandas
import sys
sys.path.append("/content/iita_python")
sys.path.append("/content/kst")
import numpy as np
import pandas as pd
import iita_python as test_iita
import learning_spaces.kst as kst
```

2. Importing the dataset, generating quasi-orders

```python
from iita_python.utils import read_rp

data_raw = read_rp("iita_python/test_ipynbs/pisa.csv")
data = test_iita.Dataset(data_raw)

unfolded_ce = test_iita.unfold_examples(data.counterexamples)
qos = test_iita.ind_gen(unfolded_ce, data.items)
```

3. Testing the fit metrics

3.1. Original IITA fit metric

```python
import iita_python.fit_metrics as test_fit
corrects = kst.iita(data.response_patterns, 3)['diff']

for i, qo in enumerate(qos):
  test_metric = round(test_fit.orig_iita_fit(data, qo), 3)
  corr_metric = round(corrects[i], 3)

  try:
    assert (test_metric == corr_metric)
    print(f'Fit metric {i}/{len(qos)} correct ✅')
  except AssertionError:
    print(f'Fit metric {i}/{len(qos)} incorrect ❌')
    print(f'Test: {test_metric}')
    print(f'Correct: {corr_metric}')
```

3.2. Corrected fit metric

```python
corrects = kst.iita(data.response_patterns, 2)['diff']

for i, qo in enumerate(qos):
  test_metric = round(test_fit.corr_iita_fit(data, qo), 3)
  corr_metric = round(corrects[i], 3)

  try:
    assert (test_metric == corr_metric)
    print(f'Fit metric {i}/{len(qos)} correct ✅')
  except AssertionError:
    print(f'Fit metric {i}/{len(qos)} incorrect ❌')
    print(f'Test: {test_metric}')
    print(f'Correct: {corr_metric}')
```
