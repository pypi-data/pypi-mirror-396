```python
# 1. Magic command (Must be in the notebook, not the module)
%matplotlib widget

import numpy as np
from volviz import VolumeSlicer 

# 2. Generate dummy data (11x11x17)
dims = [11, 11, 17]
vol1 = np.random.rand(*dims)
vol2 = np.random.rand(*dims) * 0.5
vol3 = np.random.rand(*dims) + 0.5
vol4 = np.random.rand(*dims) # This one should appear on the second row

# 3. Initialize with spacing (x=1, y=1, z=3)
slicer = VolumeSlicer(
    volumes=[vol1, vol2, vol3, vol4], 
    names=["Raw Data", "Mask", "Probability", "Overlay"],
    spacing=(1.0, 1.0, 3.0),
    cmaps=['grey', 'Blues', 'Reds', 'hot']
)

slicer.show()
```
