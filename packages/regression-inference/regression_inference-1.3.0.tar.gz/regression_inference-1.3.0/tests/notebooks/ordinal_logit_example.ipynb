```python
import pandas as pd
import numpy as np
from regression_inference import OrdinalLogisticRegression, summary
from statsmodels.miscmodels.ordinal_model import OrderedModel
import statsmodels.api as sm
```

```python
anes_data = sm.datasets.anes96.load()
anes_exog = anes_data.exog
X=pd.DataFrame(anes_exog)
y=pd.Series(anes_data.endog)
```

```python
model = OrdinalLogisticRegression().fit(X, y)
```

```python
print(model)
```

```python
model_adj = OrdinalLogisticRegression().fit(X, y, adj_cutpoints=True)
print(model_adj)
```

```python
pd.DataFrame(model.inference_table())
```

```python
pd.DataFrame(model_adj.inference_table())
```

```python
pd.DataFrame(model.predict([5, 6, 32, 4, 20], return_table=True))
```

```python
pd.DataFrame(model.predict([5, 6, 32, 4, 20], return_table=False))
```

```python
pd.DataFrame(model_adj.predict([5, 6, 32, 4, 20], return_table=True))
```

```python
pd.DataFrame(model_adj.predict([5, 6, 32, 4, 20], return_table=False))
```
