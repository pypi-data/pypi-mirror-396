```python
import pandas as pd
import numpy as np
from regression_inference import MultinomialLogisticRegression, summary
import matplotlib.pyplot as plt
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
anes_data = sm.datasets.anes96.load()
anes_exog = anes_data.exog
anes_exog = sm.add_constant(anes_exog)
X=pd.DataFrame(anes_exog)
y=pd.Series(anes_data.endog)
```

```python
model = MultinomialLogisticRegression().fit(X, y, cov_type="HC0")
model
```

```python
pd.DataFrame(model.inference_table())
```

```python
pd.DataFrame(model.variance_inflation_factor())
```

```python
print(summary(model))
```

```python
p = model.predict( np.array([[3.9, 4, 28.0, 6.0, 1]]), return_table=False )
np.sum(p)
```
