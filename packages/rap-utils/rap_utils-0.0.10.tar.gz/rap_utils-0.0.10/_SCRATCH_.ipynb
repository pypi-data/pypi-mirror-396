```python
class OneDriveClient(object):
    def __init__(self, **kwargs):
        self.client = GraphServiceClient(**kwargs)
    
    async def list_groups(self, filter=None):
        request_configuration = RequestConfiguration(
            query_parameters = GroupsRequestBuilder.GroupsRequestBuilderGetQueryParameters(
                filter=filter,
                select = ["id", "displayName"],
            )
        )

        result = await client.groups.get(request_configuration=request_configuration)

        return [(g.id, g.display_name) for g in (await client.groups.get(request_configuration=request_configuration)).value]
```
