```python
import random
import networkx as nx
import ndlib.models.ModelConfig as mc
import ndlib.models.epidemics as ec
from ndlib.viz.mpl.DiffusionTrend import DiffusionTrend
import matplotlib.pyplot as plt
```

```python
G = nx.erdos_renyi_graph(1000 , 0.1)
```

```python
model = ec.SIRModel(G)
config = mc.Configuration()
config.add_model_parameter("beta" , 0.03)
config.add_model_parameter("gamma" , 0.05)
config.add_model_parameter("Percentage Infected" , 0.25)

model.set_initial_status(config)

iterations = model.iteration_bunch(500)
trends = model.build_trends(iterations)

viz_sir = DiffusionTrend(model, trends)
viz_sir.plot(filename="AnshNewLatestTrial_SIRMODEL.png")
```

```python
model = ec.SISModel(G)
config = mc.Configuration()
config.add_model_parameter("beta" , 0.01)
config.add_model_parameter("lambda" , 0.03)
config.add_model_parameter("Percentage Infected" , 0.25)

model.set_initial_status(config)

iterations = model.iteration_bunch(200)
trends = model.build_trends(iterations)

viz_sis = DiffusionTrend(model , trends)
viz_sis.plot(filename="AnshNewLatestTrial_SISMODEL.png")
```

```python
model = ec.SEIRModel(G)

config = mc.Configuration()
config.add_model_parameter("beta" , 0.04)
config.add_model_parameter("alpha" , 0.02)
config.add_model_parameter("gamma" , 0.03)
config.add_model_parameter("Percentage Infected" , 0.25)

model.set_initial_status(config)

iterations = model.iteration_bunch(200)
trends = model.build_trends(iterations)

viz_seir = DiffusionTrend(model , trends)
viz_seir.plot(filename="AnshNewLatestTrial_SEIRMODEL.png")
```

```python
model_ic = ec.IndependentCascadesModel(G)

cfg_ic = mc.Configuration()
cfg_ic.add_model_initial_configuration("Infected", list(range(10)))
cfg_ic.add_model_parameter("threshold", 0.5)
cfg_ic.add_model_parameter("fraction_infected", 0.05)
model_ic.set_initial_status(cfg_ic)

its_ic = model_ic.iteration_bunch(200)
tr_ic = model_ic.build_trends(its_ic)

viz_ic = DiffusionTrend(model_ic, tr_ic)
viz_ic.plot(filename="ansh_ic_diffusion_trend.png")
```
